# Mitä pitikään tehdä?

- Voisiko antaa arvon a = kasvava bensankulutus litroina / voitettu aika sekunteina(tai min tai h)?
- Vaihtoehtoisesti vain absoluuttiset arvot. Sitten verrataan suuruutta.

## Sovellus tekee...
- vertailun matka-ajan ja polttoaineen kulutuksen välillä kahden eri nopeuden mukaan

### Input
- matkustus_etäisyys
- kulkuneuvotyyppi
- nopeus1
- nopeus2

### Output
- aika_nopeudella1 = nopeus1 * etäisyys
- aika_nopeudella2 = nopeus2 * etäisyys
- kulutus_nopeudella1 = 1.009**(nopeus1-1) * ajoneuvon_kulutus
- kulutus_nopeudella2 =
- aikojen_ero = voitettu_aika
- kulutusten_ero = kohonnut_bensankulutus

- voitettu_aika = abs(aika_nopeudella1 - aika_nopeudella2)
    - pitää määrittää kumpi nopeus on suurempi (ei tarvikaan)
    - laskea molemmilla nopeuksilla aika, joka matkaan kuluu ja laskea näiden itseisarvo
        - tällä tavoin ei tarvitse selvittää suuruusjärjestystä
        - pitäisikö ilmoittaa, että "nopeudella x aikaa kuluisi y vähemmän ja lisäkulut olisivat z"
            - suuruusjärjestys tarvitaan
- kohonnut_bensankulutus
    - bensankulutus kasvaa kulmakertoimella 1.009 per lisääntynyt 1 km/h
        - tarkoitetaanko lineaarista vai eksponentiaalista?
            - eksponentiaalisessa on enemmän järkeä
            - kulutus lasketaan kaavalla 1.009**(x-1), jossa x on vauhti
    - lähtöarvot ovat vauhdille 1 km/h

```python
import pandas as pd

#s, c, v1, v2 = input("Anna ajattettava matka, ajoneuvon kulutus (l per km), vauhti1 ja vauhti2 (erottele välilyönnein):").split()

# Laskee kulutuksen annetulle nopeudelle.
def con_v(v):
    return 1.009**(v-1) * c

# Etäisyys (km)
s = 160

# Kulutus (l / 100 km)
c = 3

# Nopeus1 (km / h)
v1 = 80

# Nopeus2 (km / h)
v2 = 40

# Kulutuksen (l / 100 km) laskeminen nopeudelle
con_v1 = con_v(v1)
con_v2 = con_v(v2)

# Kulutus (l) koko matkalle kummallakin nopeudella.
V1 = con_v1 * (s / 100) 
V2 = con_v2 * (s / 100)

# Laskee matkustusajan (h) annetulle nopeudelle.
time1 = s / v1
time2 = s / v2
print("Nopeudella", v1, "km/h kuluu", time1, "h ja nopeudella", v2, "km/h kuluu", time2, "h.")

# Laskee kulutuseron (l / 100 km).
con_d = abs(con_v1 - con_v2)

# Kuinka paljon ylimääräistä bensaa kuluu suuremmalla nopeudella (l)?
con_tot = s / 100 * con_d

# Laskee matkustukseen kuluvan aikaeron (h).
t_d = abs(time1 - time2)

# Kulutuseron ja säästyneen ajan suhde.
ratio1 = con_tot / t_d

# Säästyneen ajan ja kulutuseron suhde.
ratio2 = t_d / con_tot

# Vertaa nopeudet ja sijoittaa arvot uusiin muuttujiin.
if v1 < v2:
    vg = v2
    vl = v1
else:
    vg = v1
    vl = v2

# kokeile numpyä seuraavaks
df = pd.DataFrame(columns=[[v1, V1, time1],[v2, V2, time2]], index=["Nopeus", "Kulutus", "Matka-aika"],)
display(df)
    
print(f"Polttoainetta kuluu {V1:.2f} litraa nopeudella {v1} km/h ja polttoainetta kuluu {V2:.2f} litraa nopeudella {v2} km/h.")
print("Polttoainetta kuluu {:.3f} l/100 km enemmän nopeudella {} km/h, mutta aikaa säästyy {} h.".format(con_d,vg,t_d))
# print("Bensaa siis kuluu {:.2f} l enemmän matkalla {} kilometriä.".format(con_tot, s)) # Pitääkö tämä oikeasti paikkansa?
# print("Aikaa säästyy", t_d, "min.")
print(f"{ratio1:.2f} l/100 km / h")
print(f"{ratio2:.2f} h / l/100 km")
```
